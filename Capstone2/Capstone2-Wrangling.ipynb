### Cleaning 

```python
#Replace all NaN values in 'CPI' column with the mean of column values
#df['S2'].fillna(value=df['S2'].mean(), inplace=True)
features_data['CPI'].fillna(value=features_data['CPI'].mean(), inplace=True)
#Ref-https://thispointer.com/pandas-replace-nan-with-mean-or-average-in-dataframe-using-fillna/
#features_dataset_zero['CPI'].fillna(method='ffill')
```

```python
features_data['CPI'].describe()
```

```python
#Check if there are any Nan values
features_data[features_data['CPI'].isna()] # This returns an empty dataset, which means there are no missing values.
```

```python
#Replace all NaN values in 'Unemployment' column with the mean of column values
features_data['Unemployment'].fillna(value=features_data['Unemployment'].mean(), inplace=True)
##Ref:https://thispointer.com/pandas-replace-nan-with-mean-or-average-in-dataframe-using-fillna/
```
