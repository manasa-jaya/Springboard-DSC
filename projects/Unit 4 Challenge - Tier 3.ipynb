**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

##Now lets write the function to calculate price_ratio 

```python
d=dfg[dfg['London_Borough']=='Camden'] #Filter dfg and select a any row of London Borough, save this result in a variable 'd'
def create_price_ratio(d):
    y1998 = float(d['Average Price'][d['Year']==1998])
    y2018 = float(d['Average Price'][d['Year']==2018])
    ratio = [y2018/y1998]
    return ratio

```

```python
#Call the function create_price_ratio and check if it works
create_price_ratio(dfg[dfg['London_Borough']=='Camden'])
```

```python
Create an empty dictionary called all_ratios which stores ratios for each unique London Borough
```

```python
all_ratios = {}  #Create an empty dictionary 
```

```python
#iterate through each of the unque values of London_Borough column of our dataframe dfg
for b in dfg['London_Borough'].unique():
    #we subset dfg on 'London_Borough' == b. 
    borough = dfg[dfg['London_Borough'] == b]
    # create_price_ratio with the argument'borough'
    all_ratios[b] = create_price_ratio(borough) #We use the function and incorporate that into a new key of the dictionary 
print(all_ratios)  
```

##Now we got a dictionary that contains all London_Boroughs with their respective average price ratios, 
now convert it to a dataframe

```python
df_ratios = pd.DataFrame(all_ratios) #Convert all_ratios to a dataframe and save the result in df_ratios
```

```python
df_ratios.head()  #Check and see if it is converted
```

```python
#Now transpose it , and reset the index
df_ratios_T = df_ratios.transpose()
df_ratios_T.head()
```

```python
#Reset it's index and store the result
df_new = df_ratios_T.reset_index()
df_new.head()
```

```python
#Now rename the columns
df_new.rename(columns={'index':'Borough',0:'2018'},inplace=True)
df_new.head()
```

```python
#Now let's select the top15 boroughs(whose prices have increased the highest in 2 decades)
top_15 = df_new.sort_values(by='2018', ascending=False).head(15) #sELECT TOP15 in descending order
print(top_15)
#type(top_15)
```

```python
#Plot a bar garph to see the Boroughs that have seen the greatest change in prices in the past 2 decades
#Filter top_15 on 'Borough' and '2018' and plot a graph and store the result in graph_result.

graph_result = top_15[['Borough','2018']].plot(kind='bar')
#graph_result.set_xlabel('Boroughs')
graph_result.set_ylabel('Price_Ratio')
graph_result.set_xticklabels(top_15.Borough)
```
